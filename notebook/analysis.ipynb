```python
import pandas as pd
import json
print('* loaded all libs')
```

```python
# load data
df = pd.read_excel('data-slr.xlsx')

print('* loaded data', len(df))

df.head()
```

```python
# for year column, ffill
df['Year'] = df['Year'].fillna(method='ffill')

# for Year column, convert to int
df['Year'] = df['Year'].astype(int)

# fill all NaN values with null
df = df.fillna('')


# for all numeric columns, convert to int
numeric_columns = df.select_dtypes(include=['number']).columns
df[numeric_columns] = df[numeric_columns].astype(int)
```

```python
# save a JSON file, wrap the data in a 'data' key
json_data = {'data': df.to_dict(orient='records')}
with open('../web/public/data-slr.json', 'w') as f:
    json.dump(json_data, f, indent=2)

print('* saved data-slr.json')
```
